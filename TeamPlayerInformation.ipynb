```python
all_players = []
for team in team_ids:
    response = requests.get(f"http://lookup-service-prod.mlb.com//json/named.roster_40.bam?team_id={team}")
    j = response.json()
    rows = j['roster_40']['queryResults']['row']
    for row in rows:
        all_players.append(row)
```

```python
def parseFloat(val):
    try:
        return float(val)
    except:
        return 0.000
```

```python
hitting = []
total_players = len(player_ids)
player_count = 1
for player in player_ids:
    print(player)
    data = {}
    response = requests.get(f"http://lookup-service-prod.mlb.com/json/named.sport_hitting_tm.bam?league_list_id='mlb'&game_type='R'&season='2019'&player_id='{player}'")
    j = response.json()
    if int(j['sport_hitting_tm']['queryResults']['totalSize']) > 0:
        print(f"found results {player}")
        if int(j['sport_hitting_tm']['queryResults']['totalSize']) > 1:
            for i in range(int(j['sport_hitting_tm']['queryResults']['totalSize'])):
                print(f"parsing record by index {i}")
                data['player_id'] = player
                data['bb'] = data.get('bb', 0) + int(j['sport_hitting_tm']['queryResults']['row'][i]['bb'])
                data['avg'] = data.get('avg', 0.000) + parseFloat(j['sport_hitting_tm']['queryResults']['row'][i]['avg'])
                data['slg'] = data.get('slg', 0.000) + parseFloat(j['sport_hitting_tm']['queryResults']['row'][i]['slg'])
                data['ops'] = data.get('ops', 0.000) + parseFloat(j['sport_hitting_tm']['queryResults']['row'][i]['ops'])
                data['cs'] = data.get('cs', 0) + int(j['sport_hitting_tm']['queryResults']['row'][i]['cs'])
                data['sb'] = data.get('sb', 0) + int(j['sport_hitting_tm']['queryResults']['row'][i]['sb'])
                data['hr'] = data.get('hr', 0) + int(j['sport_hitting_tm']['queryResults']['row'][i]['hr'])
                data['rbi'] = data.get('rbi', 0) + int(j['sport_hitting_tm']['queryResults']['row'][i]['rbi'])
                data['lob'] = data.get('lob', 0) + int(j['sport_hitting_tm']['queryResults']['row'][i]['lob'])
                data['xbh'] = data.get('xbh', 0) + int(j['sport_hitting_tm']['queryResults']['row'][i]['xbh'])
                data['obp'] = data.get('obp', 0.000) + parseFloat(j['sport_hitting_tm']['queryResults']['row'][i]['obp'])
                data['h'] = data.get('h', 0) + int(j['sport_hitting_tm']['queryResults']['row'][i]['h'])
                data['r'] = data.get('r',0) + int(j['sport_hitting_tm']['queryResults']['row'][i]['r'])
                data['ab'] = data.get('ab', 0) + int(j['sport_hitting_tm']['queryResults']['row'][i]['ab'])
                data['so'] = data.get('so', 0) + int(j['sport_hitting_tm']['queryResults']['row'][i]['so'])
                data['babip'] = data.get('babip', 0.000) + parseFloat(j['sport_hitting_tm']['queryResults']['row'][i]['babip'])
            
            cnt = i + 1
            print(f"calculating averaged based on count: {cnt}")
            data['slg'] = data.get('slg') / cnt
            data['ops'] = data.get('ops') / cnt
            data['avg'] = data.get('avg') / cnt
            data['obp'] = data.get('obp') / cnt
            data['babip'] = data.get('babip') / cnt
            hitting.append(data)
        else: 
            data['player_id'] = player
            data['bb'] = int(j['sport_hitting_tm']['queryResults']['row']['bb'])
            data['avg'] = parseFloat(j['sport_hitting_tm']['queryResults']['row']['avg'])
            data['slg'] = parseFloat(j['sport_hitting_tm']['queryResults']['row']['slg'])
            data['ops'] = parseFloat(j['sport_hitting_tm']['queryResults']['row']['ops'])
            data['cs'] = int(j['sport_hitting_tm']['queryResults']['row']['cs'])
            data['sb'] = int(j['sport_hitting_tm']['queryResults']['row']['sb'])
            data['hr'] = int(j['sport_hitting_tm']['queryResults']['row']['hr'])
            data['rbi'] = int(j['sport_hitting_tm']['queryResults']['row']['rbi'])
            data['lob'] = int(j['sport_hitting_tm']['queryResults']['row']['lob'])
            data['xbh'] = int(j['sport_hitting_tm']['queryResults']['row']['xbh'])
            data['obp'] = parseFloat(j['sport_hitting_tm']['queryResults']['row']['obp'])
            data['h'] = int(j['sport_hitting_tm']['queryResults']['row']['h'])
            data['r'] = int(j['sport_hitting_tm']['queryResults']['row']['r'])
            data['ab'] = int(j['sport_hitting_tm']['queryResults']['row']['ab'])
            data['so'] = int(j['sport_hitting_tm']['queryResults']['row']['so'])
            data['babip'] = parseFloat(j['sport_hitting_tm']['queryResults']['row']['babip'])
            hitting.append(data)
    print(f"finished search {player_count} of {total_players}")
    player_count = player_count + 1
```
